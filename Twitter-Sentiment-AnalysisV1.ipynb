### Project 1 : Diabetes prediction

```python
#Import Keras and Numpy
from keras.models import Sequential #Import Sequential function from Keras.models
from keras.layers import Dense #Import Dense function from Keras.layers
import numpy 
numpy.random.seed(7) #Set the random seed to 7 so that pseudo random numbers generated everytime will start from the same point.
```

```python
#Load data into dataset from csv file in directory using loadtxt function from numpy
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
```

```python
# Split into X and Y as input and output variables from the imported dataset. Indices represent rows and columns to be chosen from dataset.
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
#Define Model-arguments include number of input and output layers to be built, and the activation function to be used in each layer 
model = Sequential()#used to define empty network to build the model upon
model.add(Dense(12, input_dim=8, activation='relu'))#Add 1st layer with 8 dimensional input and 12 dimensional output with Rectified Linear Unit as activation fuction  
model.add(Dense(8, activation='relu'))#Add 2nd layer with 8 dimensional output with Rectified Linear Unit as activation fuction. 12 dimensional input need not be defined expicitely as its an output from previous layer
model.add(Dense(1, activation='sigmoid'))#Add 3rd layer with 1 dimensional output with Sigmoid as activation fuction. 8 dimensional input need not be defined expicitely as its an output from previous layer
```

```python
#Compile model- to determine the loss function, optimizer and metric for evaluation of the model. Here, binary cross entropy is used as loss function since the output needs to be a binary classification of either 0 (no diabeties) or 1 (onset of diabeties), to determine gap between predicted and true label. Adam optimization algorithm is used to update parameters as its computationally efficient. Accuracy of prediction of labels is used as a metric to judge the performance of the defined model.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Model is trained by giving it inputs and corresponding labels(outputs) from dataset. Parameters are updated in mini-batches of 10 and 150 rounds of an epoch i.e an entire round of updating all parameters. Hence total number of updates are 768/10=76,76*150=11400 
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
#Model is evaluated on the dataset to determine the level of accuracy attained while predicting the labels. Here, we are evaluating the model basis the same data that is used to train it. 
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
# Accuracy attained for this model is 78.26% i.e 78 out of 100 times, this model is able to predict correct labels for given inputs. We infer that the model's accuracy is not best and can be increased by improving network structure, reconsidering the activation functions/loss function/optimizer, changing the number of epochs/batch size as some alternatives. 
```

### Project 2 : Twitter sentiment identification.

```python
import pandas as pd #import pandas module for importing data
```

```python
#Load data into dataset2 from csv file in directory using read_csv function from pandas library. 'Encoding' argument is used to import the file in compatible format. Header=0 indicates first row being headings of the dataset. 
dataset2 = pd.read_csv("train_validation.csv",encoding = "ISO-8859-1",header=0)
```

```python
#Pre-processing of text data before building the model is important as it eliminates unwanted characters and transfroms data into desired form to save further computational hassles and time
```

```python
#Tokenized text data using TreebankWordTokenizer from NLTK library, nltk.tokenize module
import nltk
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
text = dataset2.iloc[:,2] # considers 'messages' column from dataset2
token=text.apply(tokenizer.tokenize) # splits sentences into lists of words 
print(token)
print(type(token))
```

```python
#Removed punctuations, numbers and special characters using re module
import re

pattern = r'[^A-Za-z ]' #defines pattern as a non alphabetic letter 
regex = re.compile(pattern) #Compiling regular expressions into objects defined in pattern 
remove_punct = []   

for i,w in token.iteritems(): #iterating over each word in each list of tokens
    temp = [regex.sub('', word) for word in w] #.sub() function from re module used to replace any non-alphabet characters to blanks from compiled regular expressions 'regex'
    remove_punct.append(temp)
    
remove_punct
```

```python
#Converted all text to lowercase as Python is case-sensitive
lowercase1=[]
remove_punct=pd.Series(remove_punct).astype(str) #converted remove_punct to desired pandas series to be able to apply required functions
for i1,w1 in remove_punct.iteritems():
    lower=w1.lower()# converted all text to lowercase
    lowercase1.append(lower)
lowercase1
```

```python
lowercase1= pd.Series(lowercase1)#converted lowercase1 to desired pandas series to be able to apply required functions
```

```python
#Used CountVectorizer function from sklearn library for feature extraction. Vectorized each tweet to indicate occurence of words, 1 if present 0 if not.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer('stop_words=english',min_df=0.0005,max_df=0.7)#CountVectorizer arguments: in-built list of common english stopwords used to remove common words to improve computational efficiency and get rid of redundancies, min_df =0.0005 implies that infrequent words occurring in less than 0.05% of tweets are removed, max_df=0.7 implies that overused words occurring in more than 70% of tweets are removed 
X1 = vectorizer.fit_transform(lowercase1)#to build the vocabulary and encode tweets basis presence of words in the particular tweet
print(vectorizer.get_feature_names())
X1.toarray()#vectors transformed to arrays to be used as inputs to the neural network model
```

```python
print(X1.shape)#tells the number of instances captured for every tweet, 1796 here
```

```python
# Split the dataset into training and testing data with slicing to ensure first 80000 tweets in train set and remaining in test set
X_train=X1[0:80000,:]
Y_train=dataset2.iloc[0:80000,1]
X_test=X1[80000:,:]
Y_test=dataset2.iloc[80000:,1]
```

```python
model = Sequential()#used to define empty network
model.add(Dense(12, input_dim=1796, activation='relu'))#Add 1st layer with input as vectorized array each with 1796 instances and 12 dimensional output with Rectified Linear Unit as activation fuction  
model.add(Dense(8, activation='relu'))#Add 2nd layer with 8 dimensional output with Rectified Linear Unit as activation fuction. 12 dimensional input need not be defined expicitely as its an output from previous layer
model.add(Dense(8, activation='relu'))#Add 3rd layer with 8 dimensional output with Rectified Linear Unit as activation fuction. 8 dimensional input need not be defined expicitely as its an output from previous layer
model.add(Dense(8, activation='relu'))#Add 4th layer with 8 dimensional output with Rectified Linear Unit as activation fuction. 8 dimensional input need not be defined expicitely as its an output from previous layer
model.add(Dense(1, activation='sigmoid'))#Add 5th layer with 1 dimensional output with Sigmoid as activation fuction. 8 dimensional input need not be defined expicitely as its an output from previous layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #Compile model- to determine the loss function, optimizer and metric for evaluation of the model. Here, binary cross entropy is used as loss function since the output needs to be a binary classification of either 0 (negative sentiment) or 1 (positive sentiment), to determine gap between predicted and true label. Adam optimization algorithm is used to update parameters as its computationally efficient. Accuracy of prediction of labels is used as a metric to judge the performance of the defined model.
```

```python
model.fit(X_train, Y_train, epochs=100, batch_size=25000)#Model is trained by giving it inputs and corresponding labels(outputs) from dataset. Parameters are updated in batches of 25000 and 100 rounds of an epoch i.e an entire round of updating all parameters. Hence total number of updates are 80000/25000=3,3*100=300
```

```python
#Model is evaluated on the test dataset to determine the level of accuracy attained while predicting the labels. 
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
# Accuracy attained is 73.76%, this implies that 73 out of 100 times this model will correctly predict whether sentiment of the tweet is negative or positive
```
